```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
font_path = '../img/Poppins/Poppins-Regular.ttf'
df = pd.read_csv('../data/3.preprocessed/tweet_preprocessed_noNaN.csv')
all_text = " ".join(df['text_without_stopword'])
wc = WordCloud(background_color='white', font_path=font_path, width=1080, height=1080, max_words=2000,
               max_font_size=512, random_state=42, collocations=False).generate(all_text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
mask = np.array(Image.open('D:/Code Hub/Project/Omicron-Sentiment-Analysis-Indonesia/img/mask.png'))
wc = WordCloud(background_color='white', font_path=font_path, width=mask.shape[1], height=mask.shape[0], 
               max_words=2000, max_font_size=512, random_state=42, collocations=False).generate(all_text)
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
